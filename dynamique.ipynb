```python
#importation de modules utiles
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# parametres de l'equation
m = 0.4 # kg
g = 9.81 # m.s^-2
h = 25 # m
beta = 0.040 # kg.m^-1

# ecriture de la fonction caracteristique de l'ED en Y(t)
def F_Y(Y,t):
    return np.array([ Y[1] , -beta/m*np.abs(Y[1])*Y[1] - g ])

# conditions initiales
Y0 = np.array([ h , 0 ])

# temps
t = np.linspace(0,3.5,1000)

# resolution de l'ED
Y = odeint(F_Y,Y0,t)
y = Y[:,0]    
vy = Y[:,1] 
```

```python
try:
    vlim = (g*m/beta)**0.5
    tau = (m/beta/vlim)
    print("vlim = ", vlim," m/s")
    print("tau = ", tau, " s")
except:
    print("pas de vitesse limite")
```

```python
# graphique de v(t)
plt.plot(t,vy,'-')
plt.xlabel('t en s')
plt.ylabel('v(t) en m/s')
plt.show()
```

```python
# graphique de y(t)
plt.plot(t,y,'-')
plt.xlabel('t en s')
plt.ylabel('y(t) en m')
plt.ylim(ymin=0)
plt.show()
```
